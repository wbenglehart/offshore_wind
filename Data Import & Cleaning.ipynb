## Data Cleaning:

### Cost and Performance Dataframe:

```python
cap_df.columns = cap_df.iloc[0]
cap_df.drop(cap_df.index[[0]], inplace = True)
cap_df.reset_index(inplace = True)
cap_df.rename(columns={'index':'Year'}, inplace=True)
cap_df.to_csv('data/cap.csv', index=False)
cap_df = pd.read_csv('data/cap.csv')
cap_df.set_index('Year', inplace = True)
cap_df.head()
```

Removing all irrelevant data from the dataframe. Though the mjaority of the features are important in relation to the problem statement being addressed, certain ones are made irrelvent by the multi-national approach. 

Grid Connection Cost is a product of that associated with U.S. market and does not include that of Canada or Mexico and so will be removed. Dropping 2019, 2020, 2021 data as as well.

```python
cap_df.drop(cap_df.iloc[:, 324:366], inplace = True, axis = 1) # Grid Connection Costs

cap_df.drop(cap_df.iloc[:, 282:324], inplace = True, axis = 1) #VOME

cap_df.drop(cap_df.index[[0,1,2]], inplace = True) #2019, 2020, 2021

cap_df.head()
```

Removing special characters and changing to numerical objects

```python
cap_df = cap_df.replace(r'[^\w\s]|_', '', regex=True)

cap_df = cap_df.astype(float)

cap_df.iloc[:, 0:72] = cap_df.iloc[:, 0:72] / 10

cap_df.to_csv('data/cap_cleaned.csv', index=False)

cap_df.head()
```

### Offshore Wind Farm 2027 Analysis Dataframe:

Setting the index as wind farm id and removing unneed factors

```python
pow_df.set_index('wfarm_id', inplace = True)

pow_df.sort_index(inplace = True)

pow_df = pow_df[~pow_df.index.duplicated(keep='first')]
```

```python
pow_df.drop(['reeds_region','State', 'LACE'], axis = 1, inplace=True)
```

```python
def clean_wind_low(cell):
    return float(cell[0].replace(u'\u200b', '').strip())

pow_df['Wind speed'].str.split('-').map(clean_wind_low);

def clean_wind_high(cell):
    return float(cell[1].replace(u'\u200b', '').strip())
  
pow_df['Wind speed'].str.split('-').map(clean_wind_high)

pow_df['Wind speed'] = (pow_df['Wind speed'].str.split('-').map(clean_wind_low) + pow_df['Wind speed'].str.split('-').map(clean_wind_high))/2
```

```python
pow_df.reset_index(inplace = True)
pow_df.drop(['wfarm_id'], axis = 1, inplace = True)
```

```python
pow_df.isnull().sum()
```

```python
pow_df.dropna(inplace = True)
```

```python
pow_df['score'] = pow_df['NCF'] * pow_df['Net value']
```

```python
new_columns_dict = {
    'Longitude': 'longitude',
    'Longitude': 'longitude',
    'Latitude': 'latitude',
    'Depth': 'depth',
    'Dist_S_to_L': 'dist_to_land',
    'Hs_Average': 'avg_wave_height',
    'NCF': 'ncf',
    'LCOE': 'lcoe',
    'type_min_lcoe': 'substructure',
    'capacity': 'capacity',
    'isfixed': 'fixed_bottom',
    'Net value': 'net_value',
    'Wind speed': 'avg_wind_speed'
}
    
pow_df.rename(columns = new_columns_dict, inplace=True)
```

```python
pow_df.to_csv('data/pow_cleaned.csv', index=False)
```
